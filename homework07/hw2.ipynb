```r
df <- read.csv('howpop_train.csv', stringsAsFactors=TRUE)
head(df)
```

# 1. В каком месяце (и какого года) было больше всего публикаций?

```r
year = table(substring(df$published, 1, 4))
```

```r
barplot(year, xlab = 'year', ylab = 'Count')
```

```r
month = table(substring(df$published, 6, 7))
```

```r
barplot(month, xlab = 'month', ylab = 'Count')
```


# 2. Проанализируйте публикации в месяце из предыдущего вопроса

```r
library(lubridate)
df['dayofweek'] <- ifelse(wday(ymd_hms(df$published)) == 1, 7, wday(ymd_hms(df$published))-1)
```

```r
tbl <- table(df$domain, df$dayofweek)
```

```r
barplot(tbl, beside = T, xlab = 'dayofweek', ylab = 'Count', legend.text = TRUE)
```

# 3. Когда лучше всего публиковать статью?

```r
df['hour'] <- substring(df$published, 12, 13)
```

```r
library(ggplot2)
ggplot(df, aes(hour, views)) + stat_summary(fun.y = sum, geom = "bar", color="blue", fill="white")
```

```r
ggplot(df, aes(hour, comments)) + stat_summary(fun.y = sum, geom = "bar", color="blue", fill="white")
```

```r
ggplot(df[df$domain=='geektimes.ru', ], aes(y = comments, x = hour)) + geom_point()
```

```r
ggplot(df[df$domain=='habrahabr.ru', ], aes(hour, comments)) + stat_summary(fun.y = sum, geom = "bar", color="blue", fill="white")
```
